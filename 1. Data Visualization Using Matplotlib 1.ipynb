```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Draw A Line Plot

```python
x=[1,2,3,7]
y=[1,7,9,8]
plt.plot(x,y)
plt.show()
```

#### Give Title, Xlabel and Ylabel 

```python
x=[1,2,3,7]
y=[1,7,9,8]
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

#### Change Font Size of Title, Xlabel and Ylabel And Also Change location

```python
x=[1,2,3,7]
y=[1,7,9,8]
plt.plot(x,y)
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.show()
```

#### Change Color & Width of The Line

```python
x=[1,2,3,7]
y=[1,7,9,8]
plt.plot(x,y,color='r',linewidth=4)
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.show()
```

#### Change Marker, Marker Size &  Line Style

```python
x=[1,2,3,7]
y=[1,7,9,8]
plt.plot(x,y,color='r',linewidth=4,marker='.',markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',
        linestyle=':')
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.show()
```

```python
x=[1,2,3,7]
y=[1,7,9,8]
plt.plot(x,y,'r:',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue')
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.show()
```

#### Add Legend To Your Graph Also Change Location

```python
x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,y,'r--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue')
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.legend(["A simple line plot"],loc=0)
plt.show()
```

#### Create Separate Lines For x and y And Change Plot Style

```python
plt.style.available
plt.style.use('seaborn')

x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,'r--<',linewidth=4,markersize=10,
        markerfacecolor='red',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.legend()
plt.show()
```

#### Xlim And Ylim

```python
plt.style.available
plt.style.use('seaborn')

x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,'r--<',linewidth=4,markersize=10,
        markerfacecolor='red',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.xlim(0,20)
plt.ylim(0,15)
plt.legend()
plt.show()
```

#### Change X and Y limit Using axis Method

```python
x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,'r--<',linewidth=4,markersize=10,
        markerfacecolor='red',markeredgecolor='blue')
plt.plot(y,'b--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue')
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
# xmin, xmax, ymin, ymax 
plt.axis([0,20,0,15])
plt.legend(["A simple line plot"],loc=0)
plt.show()
```

#### Change X and Y limit Using Xticks And Yticks.

```python
x_ticks=[x for x in range(0,20,2)]
y_ticks=[y for y in range(0,15,3)]
x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,'r--<',linewidth=4,markersize=10,
        markerfacecolor='red',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.show()
```

#### Use grid method

```python
x_ticks=[x for x in range(0,20,2)]
y_ticks=[y for y in range(0,15,3)]
x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,'r--<',linewidth=4,markersize=10,
        markerfacecolor='red',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.grid()
plt.show()
```

#### Change figure size

```python
x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.figure(figsize=(10,2), facecolor = 'lightgreen')
plt.plot(x,'r--<',linewidth=4,markersize=10,
        markerfacecolor='red',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.grid(color = 'b', linestyle = '--', linewidth = '2')
plt.show()
```

#### Save Graph

```python
x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.figure(figsize=(10,2), facecolor = 'lightgreen')
plt.plot(x,'r--<',linewidth=4,markersize=10,
        markerfacecolor='red',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.grid(color = 'b', linestyle = '--', linewidth = '2')
plt.savefig('lineplot.png')
plt.show()
```
